```python
import pandas as pd

bond = pd.read_csv("jamesbond.csv", index_col="Film")
bond.sort_index(inplace = True)

bond.head()
```

```python
bond.sample()     # Random Sample asbe
```

```python
bond.sample(n = 5)   # n = joto gula Random degit lagbo
```

```python
bond.sample(frac=.25)   # frac = koto parcent Sample lagbo data set theke
```

```python
bond.sample(n = 3, axis="columns")   # 3 ta column er Random Sample Create hbe
```

```python
bond.sample(n = 3, axis="rows")   # 3 ta Row er Random Sample Create hbe
```
