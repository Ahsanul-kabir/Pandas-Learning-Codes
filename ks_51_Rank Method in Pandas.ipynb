```python
import pandas as pd

nba = pd.read_csv("nba.csv").dropna(how="all")

nba["Salary"] = nba["Salary"].fillna(0).astype(int)   # 'astype'  diya float ke intiger e neya hoise

nba.head()
```

```python
nba["Salary"].rank(ascending=False).astype(int)     # ulta palta astese
```

```python
nba["SalaryRank"] = nba["Salary"].rank(ascending=False).astype("int")   # ulta palta astese 
                    # Rank mane dui ta column er modde ekta column ke base raikha onno ta ke...Sort kora
nba.head()
```

```python
nba["SalaryRank"] = nba["Salary"].rank(ascending=False).astype("int") # Order "False" dite hbe ...>> boro number upor e thak te hbe

nba.sort_values(by ="Salary", ascending=False)    # Order "False" bcoz Salary er sapekkhe ranking kortesi
```
