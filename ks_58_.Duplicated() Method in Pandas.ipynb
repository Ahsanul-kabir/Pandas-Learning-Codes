```python
import pandas as pd

df = pd.read_csv("employees.csv", parse_dates=["Start Date","Last Login Time"])
df["Senior Management"] = df["Senior Management"].astype("bool")
df["Gender"] = df["Gender"].astype("category")
df.sort_values("First Name", inplace=True)
df.head()
```

```python
df["First Name"].duplicated()    # Fist value new tai "False"...porer repeat thakle old tai "True"
```

```python
df[df["First Name"].duplicated()].head()   # jei gula "True" oi gula print korbo sudu
```

```python
df["First Name"].duplicated(keep = 'last')   # deplicate last theke check hbe...>> tai last theke e aage jei ta oi ta "False"..pore "True" pabo 
                # keep = 'first' ---> By default
```

```python
df[df["First Name"].duplicated(keep = 'last')].head(20)
```

```python
df["First Name"].duplicated(keep = False).head(20)  # default -->> keep = 'False'---->> keep "True" kaj kore na wrong...
```

```python
df[df["First Name"].duplicated(keep = False)].head()
```

```python
mask = ~df["First Name"].duplicated(keep = False)   # unique value print "~  == tril diya"

df[mask]
```
